<H5>
    USE THE K-MEANS UNSUPERVISED CLUSTERING ALGORITHM TO CLASSIFY FLOWERS BASED ON THEIR FEATURES
</H5>

```python
#The number of our target is 3 ['setosa' 'versicolor' 'virginica']
model = KMeans(n_clusters=3)
```

```python
#Fitting the model
model.fit(iris_df.data)
```

```python
#Predicting a single input of the four features 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'
predicted_label = model.predict([[7.2, 3.5, 0.8, 1.6]])
print(predicted_label)
```

```python
#Predict on the entire dataset
all_predictions = model.predict(iris_df.data)
```
